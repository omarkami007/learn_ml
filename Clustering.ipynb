### **K-Means Clustering**

**K-Means Clustering** is an **unsupervised learning algorithm** used to group a dataset into **$K$ distinct clusters** based on the similarity of data points.

---

#### **Algorithm**

##### **Step 1: Initialize Centroids**
- Randomly select $K$ points from the dataset as the **initial centroids** (means of the clusters).
- These centroids are the center points around which the clusters will be formed.

##### **Step 2: Assign Points to Clusters**
- For each data point $x_i$, calculate its distance to each centroid using a distance metric (usually **Euclidean distance**):

- Assign $x_i$ to the cluster of the nearest centroid.

##### **Step 3: Update Centroids**
- Recalculate the centroids by taking the **mean of all points** assigned to each cluster:
  $
  c_j = \frac{1}{|C_j|} \sum_{x_i \in C_j} x_i
  $
  Where:
  - $C_j$: The set of points in cluster $j$.
  - $|C_j|$: The number of points in cluster $j$.

##### **Step 4: Repeat Until Convergence**
- Repeat **Steps 2 and 3** until the cluster assignments do not change or the centroids stabilize.

---

#### **The Objective Function**

The objective of K-Means is to minimize the **Within-Cluster Sum of Squares (WCSS)**, also known as the **inertia**:
$
\text{WCSS} = \sum_{j=1}^K \sum_{x_i \in C_j} \|x_i - c_j\|^2
$
- $\|x_i - c_j\|^2$: The squared distance between a point $x_i$ and its cluster centroid $c_j$.
---

#### **4. Practical Considerations**

##### **4.1 Choosing the Number of Clusters ($K$)**
One common method to determine $K$ is the **Elbow Method**:
- Plot the WCSS for different values of $K$.
- Look for the "elbow" point where the rate of decrease in WCSS slows down. This is often the optimal $K$.

##### **4.2 Local minimum**
Each of steps 2 & 3 reduces the value of the criterion, so that
 convergence is assured. However, the result may represent a suboptimal
 local minimum. The algorithm of Hartigan and Wong (1979) goes further,
 and ensures that there is no single switch of an observation from one group
 to another group that will decrease the objective. In addition, **one should
 start the algorithm with many different random choices for the starting
 means, and choose the solution having smallest value of the objective func
tion**.




```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis', s=50)
centroids = kmeans.cluster_centers_
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.legend()
plt.title('K-Means Clustering')
plt.show()
```
